```python
def getPoster(path):
    try:
        base_url = 'http://image.tmdb.org/t/p/'
        psize = 'original'
        if path != None:
            urllib.request.urlretrieve(base_url+psize+path, poster_path + path)
    except Exception as e:
        print("    **Failed to download")
```

### To prepare CSV

```python
from tmdbv3api import Movie
import urllib.request
import socket
from os.path import exists as file_exists

# Set the default timeout in seconds
timeout = 20
socket.setdefaulttimeout(timeout)

movie = Movie()
movie.language='en-US'
i=1
mid_list = []
title_list = []
post_path_list = []
genres_list = []
errorList = []
allPostersPath = '/Users/shiv/Desktop/NYU/sem3_fall/CV/Project/archive/resizedOP'
for mid in moviesID:
    try:
        mjson = movie.details(mid)
        print(i ,  mjson.id)
#         print(allPostersPath+mjson.poster_path[0:])
        
        if(file_exists(allPostersPath+mjson.poster_path[0:]) and mjson.genres is not None and mjson.title is not None):
            mid_list.append(mid)
            title_list.append(mjson.title)
            post_path_list.append(mjson.poster_path)
            genres_list.append(mjson.genres)
#         print(mjson.genres)
#         if(i>=10):
#             break
        i+=1
    except Exception as e:
        errorList.append(mid)
        print("  **Error", i)
```
